```python
# Answer to the question
import numpy as np
import matplotlib.pyplot as plt

sizex = 1000
x = np.linspace(0,10,sizex)
y = 3*x+5

# Create a matrix of random perturbations
# The shape of the matrix is 100 (number of x values) x 9 (number of perturbations)

# Créer une matrice de perturbations aléatoires
# La forme de la matrice est 100 (nombre de valeurs x) x 9 (nombre de perturbations)

perturbations = np.random.normal(0, 2, size=(sizex, 9))

# Create the perturbed y values
# y[:, np.newaxis] changes y into a column vector so that broadcasting can work
# y_pert will have the same shape as perturbations, i.e., (100, 9)

# Créer les valeurs y perturbées
# y[:, np.newaxis] change y en vecteur colonne pour que la diffusion puisse fonctionner
# y_pert aura la même forme que les perturbations, c'est-à-dire (100, 9)

y_pert = y[:, np.newaxis] + perturbations

# Calculate the mean along axis=1, which will produce a 1D array of length 100

# Calculez la moyenne le long de l'axe = 1, ce qui produira un tableau 1D de longueur 100

y_pert_mean = np.mean(y_pert, axis=1)


# Calculate the standard deviation along axis=1 for y_pert
# Calculer l'écart type le long de l'axe = 1 pour y_pert
y_pert_std = np.std(y_pert, axis=1)

# Create the plot
# Créer l'intrigue
plt.figure(figsize=(10, 6))

# Plot the original y line and the perturbed y
# Tracez la ligne y d'origine et y perturbé
plt.plot(x, y, label='y = 3x + 5', color='black')
plt.plot(x,y_pert,marker=".",linestyle="None",color="red")

# Plot the mean of y_pert
# Tracez la moyenne de y_pert
plt.plot(x, y_pert_mean, label='Mean of y_pert', color='blue')

# Add shaded areas to represent the error (standard deviation)
# Ajoutez des zones ombrées pour représenter l'erreur (écart type)
plt.fill_between(x, y_pert_mean - y_pert_std, y_pert_mean + y_pert_std, color='blue', alpha=0.2, label='Error (1 std. dev.)')

# Add labels and title
# Ajouter des étiquettes et un titre
plt.xlabel('x')
plt.ylabel('y and y_pert')
plt.title('Original Line and Perturbed Data')
plt.legend()

# Show the plot
# Afficher l'intrigue
plt.show()

```
