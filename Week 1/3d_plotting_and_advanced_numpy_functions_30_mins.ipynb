## 1. 3D Plotting

Matplotlib provides tools to create a variety of 3D plots.

### Surface Plots

Surface plots are used to display three-dimensional data.

### Contour Plots

Contour plots are like a 3D surface plot, but without the faces.

Let's explore some 3D plotting techniques.


````python

import matplotlib.pyplot as plt
import numpy as np

# Create data for 3D plotting
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Surface Plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_title('Surface Plot')
plt.show()

# Contour Plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.contour(X, Y, Z, 50, cmap='viridis')
ax.set_title('Contour Plot')
plt.show()
````


## 2. Advanced Numpy Functions

Numpy provides a range of advanced mathematical and statistical functions that are very useful for scientific computing.

### Mathematical Functions

Functions like `sin`, `cos`, `exp`, etc.

### Statistical Functions

Functions like `mean`, `std`, `min`, `max`, etc.

Let's explore some of these functions.


````python

# Mathematical Functions
arr = np.array([0, np.pi/4, np.pi/2])
print('sin:', np.sin(arr))
print('cos:', np.cos(arr))
print('exp:', np.exp(arr))

# Statistical Functions
arr = np.array([1, 2, 3, 4, 5])
print('Mean:', np.mean(arr))
print('Standard Deviation:', np.std(arr))
print('Min:', np.min(arr))
print('Max:', np.max(arr))
````
