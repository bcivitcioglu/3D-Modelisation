## 1. Multiple Plots in One Figure

Matplotlib allows for multiple plots in a single figure, which is useful for comparing different datasets or visualizations.

### Subplots

Subplots divide a figure into a grid of subplots.

### Twin Axes

Twin axes are useful when you want to show two different variables with different scales on the same plot.

Let's demonstrate these techniques.


````python

import matplotlib.pyplot as plt
import numpy as np

# Create some data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Subplots
fig, axs = plt.subplots(1, 2, figsize=(14, 6))
axs[0].plot(x, y1, 'r--')
axs[0].set_title('Sine Curve')
axs[1].plot(x, y2, 'g--')
axs[1].set_title('Cosine Curve')

# Twin Axes
fig, ax1 = plt.subplots(figsize=(7, 6))
ax2 = ax1.twinx()
ax1.plot(x, y1, 'r--')
ax1.set_xlabel('x')
ax1.set_ylabel('Sine', color='tab:red')
ax2.plot(x, y2, 'g--')
ax2.set_ylabel('Cosine', color='tab:green')

plt.show()
````


## 2. Histograms and Bar Charts

### Histograms

Histograms are used for frequency distribution representation.

### Bar Charts

Bar charts are used for comparing quantities of different categories.

Let's see how to make these plots.


````python

# Histogram
data = np.random.randn(1000)
plt.figure(figsize=(7, 6))
plt.hist(data, bins=30, alpha=0.7, color='blue', edgecolor='black')
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()

# Bar Chart
categories = ['A', 'B', 'C']
values = [3, 7, 1]
plt.figure(figsize=(7, 6))
plt.bar(categories, values, color=['red', 'green', 'blue'])
plt.title('Bar Chart')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
````


## 3. Error Bars and Filled Plots

### Error Bars

Error bars are used to show the variability in the data.

### Filled Plots

Filled plots are used to show area between lines.

Let's see how to make these plots.


````python

# Data
x = np.linspace(0, 10, 50)
y = np.sin(x)
y_err = 0.1

# Error Bars
plt.figure(figsize=(7, 6))
plt.errorbar(x, y, yerr=y_err, fmt='o-', label='Data with Error Bars')
plt.title('Error Bars')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()

# Filled Plot
y_upper = y + y_err
y_lower = y - y_err

plt.figure(figsize=(7, 6))
plt.plot(x, y, label='Data')
plt.fill_between(x, y_lower, y_upper, alpha=0.2, label='Error')
plt.title('Filled Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
````
