```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([0,1,2,3,4,5])
print(x)
```

```python
x = np.arange(6)
print(x)
```

```python
y = np.array([0 ,1,4,9,16,25])
print(y)
```

```python
y = x**2
```

```python
plt.plot(x,y)
```

```python
# Create figure
plt.figure(figsize=(12,4))

# Create the plot
plt.plot(x,y)

# Add spesifications
plt.xlabel("x")
plt.ylabel("y")
plt.title("Example Plot")
```

```python
# Create figure
plt.figure(figsize=(12,4))

# Create the plot
plt.plot(x,y, marker = "o", linestyle="--")

# Add spesifications
plt.xlabel("x")
plt.ylabel("y")
plt.title("Example Plot")
```

```python
fig, ax = plt.subplots(2,1)
ax[0].plot(x,y,marker="x",linestyle="-")
ax[1].plot(y,x,marker=",",linestyle=":")
```

```python
fig, ax = plt.subplots(3,8,figsize=(12,12))
```

```python
fig, ax = plt.subplots(2,1)
ax[0].plot(x,y,marker="x",linestyle="-",color="tab:blue",label="$x^2$")
ax[0].legend()
ax[1].plot(y,x,marker=",",linestyle=":",color="tab:red",label="$x^{1/2}$")
ax[1].legend()
```

```python
fig, ax = plt.subplots(2,1)
ax[0].plot(x,y,marker="x",linestyle="-",color="tab:blue",label="$x^2$")
ax[0].legend()
ax[1].scatter(y,x,marker="o",linestyle=":",facecolors='none', edgecolors='r',s=200,label="$x^{1/2}$")
ax[1].legend()
```

```python
x = np.linspace(0,10,100)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
fig, axs = plt.subplots(1,2,figsize=(12,6))
axs[0].scatter(x,y1)
axs[1].scatter(x,y2)
```

```python
fig, axs = plt.subplots(1,2,figsize=(12,6))
axs[0].scatter(x,y1,color="tab:red",marker=".")
axs[1].scatter(x,y2,color="tab:blue",marker="*")
```

```python
from matplotlib import gridspec
# Plot figure with subplots of different sizes

fig = plt.figure(figsize=(7,7))
gridspec.GridSpec(2,2)

ax1 = plt.subplot2grid((2,2), (0,0))

ax2=plt.subplot2grid((2,2), (0,1))

ax3=plt.subplot2grid((2,2), (1,0), colspan=2, rowspan=1)

ax1.scatter(x,y1,color="tab:red",marker=".")
ax2.scatter(x,y2,color="tab:blue",marker="*")

ax3.scatter(x,y1,color="tab:red",marker=".")
ax3.scatter(x,y2,color="tab:blue",marker="*")

```
